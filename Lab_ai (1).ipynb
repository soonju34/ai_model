```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
base_dir = "/content/drive/MyDrive/Lab_ai"
img_path = os.path.join(base_dir, "cat.1.jpg")
image = Image.open(img_path)

plt.imshow(image)
plt.show()
```

```python
gray_image = image.convert('L')
gray_image_numpy = np.array(gray_image, 'uint8')

plt.imshow(gray_image_numpy, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
file_path1 = os.path.join(base_dir, "gray_cat.1.jpg")
cv2.imwrite(file_path1, gray_image_numpy)
```

```python
file_path2 = os.path.join(base_dir, "gray_image_numpy.npy")
np.save(file_path2, gray_image_numpy)
```
